# Submitting & Reading Attestations from the OLI Label Pool

### Task 0: install [OLI pip package](https://pypi.org/project/oli-python/)

`pip install oli-python`

### Task 1: create an OLI compliant label

```python
# Understanding OLI Label Schema: https://www.openlabelsinitiative.org/docs?section=label-schema
# All OLI compliant tag_ids: https://www.openlabelsinitiative.org/docs?section=tag-documentation

address = "0xC36442b4a4522E871399CD717aBDD847Ab11FE88"
chain_id = "eip155:1" # Ethereum Mainnet
tag_id = {
    "contract_name": "NonFungiblePositionManager",
    "owner_project": "uniswap",
    "usage_category": "dex"
}
```

### Task 2: submit label to OLI Label Pool

```python
from oli import OLI

pk = "your_private_key_here" # any EVM private key wallet (for reputation purposes only no funds are needed)

oli = OLI(private_key=pk)
```

```python
oli.validate_label(address, chain_id, tag_id)
```

```python
oli.submit_label(address, chain_id, tag_id)
```

### Task 3: read in your label from the OLI Label Pool

```python
oli.get_attestations()
```
